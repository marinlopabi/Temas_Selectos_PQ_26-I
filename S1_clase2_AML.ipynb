# Operadores

Son símbolos especiales que permiten realizar operaciones con variables y valores.

Operadores aritméticos

| **Operación** | **Descripción** |
|----------------|----------------|
| `a + b` | Suma `a` y `b`. |
| `a - b` | Resta `b` de `a`. |
| `a * b` | Multiplica `a` por `b`. |
| `a / b` | Divide `a` por `b`. |
| `a // b` | División entera de `a` entre `b`. Calcula el cociente sin tener en cuenta el resto. |
| `a % b` | Devuelve el módulo (residuo) de la división de `a` entre `b`. |
| `a ** b` | Eleva `a` a la potencia de `b`. |
| `-a` | Cambia el signo de `a` (negación aritmética). |
| `+a` | Devuelve el valor positivo de `a`. |


Ejemplos:

```python
10 / 3
```

```python
10 // 3 #División entera
```

```python
10 % 3 #Módulo (Residuo)
```

```python
2**3*(9+123)
```

Operadores relacionales o de comparación.


Comparan valores y devuelven True o False.


| **Operación** | **Descripción** |
|----------------|----------------|
| `a == b` | `True` si `a` es igual a `b`. |
| `a != b` | `True` si `a` no es igual a `b`. |
| `a < b`, `a <= b` | `True` si `a` es menor (o menor o igual) que `b`. |
| `a > b`, `a >= b` | `True` si `a` es mayor (o mayor o igual) que `b`. |


**Ejemplos**

```python
a = 7
b = 10

a > b
```

Operadores lógicos
| Operador | Significado | Ejemplo             |
| -------- | ----------- | ------------------- |
| `and`    | Y lógico    | `T > 300 and P > 5` |
| `or`     | O lógico    | `T > 300 or P > 5`  |
| `not`    | Negación    | `not (T > 300)`     |




- Se usan para combinar condiciones.

- Estos operadores siempre devuelven valores booleanos: True o False.

| **Operación** | **Descripción** |
|----------------|----------------|
| `a & b` | `True` si tanto `a` como `b` son `True`

Uso de operadores

**1. Ley de los gases ideales**
$$PV=nRT$$

Datos los siguientes valores obtenga la presión del gas.

n = 2.0          mol

R = 0.0821       L·atm/mol·K

T = 298           K

V = 10            L

```python

n = 2.0          # mol
R = 0.0821       # L·atm/mol·K
T = 298          # K
V = 10           # L

P = (n * R * T) / V
print(f"Presión del gas: {P:.2f} atm")
```

**2. Ecuación de Antoine (presión de vapor)**

Calcula la presión de vapor de una sustancia.

$$log_{10}(P)=A−\frac{B}{C+T}$$

Constantes de Antoine para el agua (T en °C, P en mmHg)

A = 8.07131

B= 1730.63

C= 233.426

T = 80  °C


```python
# Constantes de Antoine para el agua (T en °C, P en mmHg)
A, B, C = 8.07131, 1730.63, 233.426
T = 80  # °C

P = 10 ** (A - (B / (C + T)))
print(f"Presión de vapor del agua a {T}°C: {P:.2f} mmHg")
```

**3. Cálculo de número de Reynolds (flujo de fluidos)**

El número de Reynolds Re permite saber si el flujo es laminar, de transición o turbulento.

$$Re=\frac{\rho vD}{μ}$$

rho = 1000    densidad [kg/m³]

v = 1.2       velocidad [m/s]

D = 0.05      diámetro [m]

mu = 0.001    viscosidad [Pa·s]




```python
rho = 1000   # densidad [kg/m³]
v = 1.2      # velocidad [m/s]
D = 0.05     # diámetro [m]
mu = 0.001   # viscosidad [Pa·s]

Re = (rho * v * D) / mu
print(f"Número de Reynolds = {Re:.0f}")
```

# Control de Flujo

Python tiene varias palabras clave internas para lógica condicional, bucles y otros conceptos estándares de control de flujo, que pueden encontrarse en otros lenguajes de programación.

#Condicionales
## if, elif y else
Permiten que el programa tome decisiones según condiciones

La sentencia **if** es uno de los tipos de sentencia de control de flujo más conocidos. Comprueba una condición, se ejecuta si la condición es verdadera.

```python
x = -56

if x < 0:
    print('x es negativo')
```

if → se ejecuta si la condición es verdadera.

elif → se ejecuta si la anterior fue falsa, pero esta verdadera.

else → se ejecuta si ninguna condición anterior fue verdadera. No lleva condición

Una sentencia *if* puede ir seguida de manera opcional por uno o más bloques **elif** y un bloque multifuncional **else** si todas las condiciones son False:

```python
x = 18
```

```python
if x < 0:
  print('x es negativo') #sangría 4 espacios
elif x==0:
    print('x es igual a cero')
else:
    print('x es positivo ')
```

Ejemplo con input

```python
n = int(input("Ingresa un número: "))

if n > 0:
    print("El número es positivo")
elif n < 0:
    print("El número es negativo")
else:
    print("El número es cero")
```
