**Ecuación cuadrática**

Dada una ecuación cuadrática $$ax^2 + bx+c=0,$$

El discriminante (Δ = b² - 4ac) determina el tipo de raíces:

Δ > 0 → Dos raíces reales distintas

Δ = 0 → Una raíz real doble

Δ < 0 → Dos raíces complejas conjugadas

Escriba un código que:
indique el tipo de raíces e emprima el valor si son reales.

```python
a=1
b=4
c=5
```

```python
D = b**2 - 4*a*c

if D > 0:
    x1 = (-b + D**0.5) / (2*a)
    x2 = (-b - D**0.5) / (2*a)
    print("Dos raíces reales distintas:", x1, x2)

elif D == 0:
    x = -b / (2*a)
    print("Una raíz real doble:", x)

else:
    print("Raíces complejas conjugadas")
```

 Verificación de conversión de una reacción. En un reactor, la conversión X indica qué tan completa está la reacción:

$$X=\frac{FA_0-FA}{FA_0}$$

```python
FA0 = 10.0   # mol/s (flujo inicial de A)
FA = 2.0     # mol/s (flujo final de A)

X = (FA0 - FA) / FA0 * 100
print(f"Conversión: {X:.1f}%")

if X < 60:
    print("Conversión baja: ajustar condiciones del reactor.")
elif 60 <= X < 90:
    print("Conversión moderada: proceso aceptable.")
else:
    print("Conversión alta: operación eficiente.")
```

# **BUCLES**

Un **bucle** sirve para **repetir instrucciones** varias veces sin tener que escribir el mismo código una y otra vez.  

En Python, los bucles más comunes son **`for`** y **`while`**.




Un bucle "le dice al programa" que repita una acción mientras se cumpla una condición o durante cierto número de veces.









##  Bucle `for`

Se usa cuando **sabemos cuántas veces** queremos repetir algo o cuando recorremos una **lista**.

```python
for i in range(5):
    print("número:", i)
```

```python
L = [1, 2, None, 4, None, 5]
```

```python
for i in L:
    print(i)
```

```python
# Suma de los primeros 1000 números

suma = 0
for i in range(1, 1001):
    suma += i

suma
```


 ## Bucle `while`


Se usa cuando no sabemos cuántas veces se repetirá,
solo sabemos que queremos repetir mientras se cumpla una condición.

```python
contador = 1 #El bucle se repite mientras la condición sea verdadera.

while contador <= 5:
    print(contador)
    contador = contador + 1
```

```python
x = 0
while x < 5:
    print("x vale:", x)
    x += 1
```

## `break` `continue`

Romper o saltar bucles

| Instrucción | Función                                                              |
| ----------- | -------------------------------------------------------------------- |
| `break`     | Detiene el bucle antes de que termine                                |
| `continue`  | Salta a la siguiente iteración sin ejecutar lo que sigue en el ciclo |


```python
for i in range(1, 10):
    if i == 5:   #Detiene el ciclo al llegar a 5
        break
    print(i)
```


| Tipo de bucle | Cuándo usarlo                                           | Ejemplo típico                    |
| ------------- | ------------------------------------------------------- | --------------------------------- |
| `for`         | Cuando sabes **cuántas veces** repetir                  | recorrer listas, rangos o cadenas |
| `while`       | Cuando repites **mientras una condición sea verdadera** | simulaciones, procesos iterativos |
| `break`       | Para **salir** del bucle antes de tiempo                | detener una simulación            |

## Ejercicio: Monitoreo de temperatura en un reactor

En un reactor continuo se mide la **temperatura de operación** cada cierto tiempo.  
Por seguridad, el reactor **no debe operar por encima de 80 °C**.

El operador registra temperaturas hasta que el proceso se detiene.

Usar **bucles**, **condicionales** y **listas** para:
- registrar datos experimentales,
- analizarlos,
- y tomar decisiones de operación.



Escribe un programa en Python que haga lo siguiente:

1. Pida al usuario temperaturas del reactor (en °C) una por una.
2. Guarde cada temperatura en una **lista**.
3. El programa debe seguir pidiendo valores **mientras la temperatura sea menor o igual a 80 °C**.
4. Si el usuario introduce una temperatura **mayor a 80 °C**, el proceso se detiene y:
   - se muestra un mensaje de **alarma**.
   



```python
# Lista donde se almacenarán las temperaturas válidas
temperaturas = []

# Inicializamos la variable temperatura
temperatura = 0

# Bucle while: se ejecuta mientras la temperatura sea segura
while temperatura <= 80:

    # Se pide la temperatura al usuario
    temperatura = float(input("Ingrese la temperatura del reactor (°C): "))

    # Condición de seguridad del proceso
    if temperatura <= 80:
        temperaturas.append(temperatura)  # Se guarda la temperatura
    else:
        print("ALARMA: Temperatura fuera del rango seguro.")
        print("Proceso detenido.")
print("\nTemperaturas registradas:", temperaturas)
```
